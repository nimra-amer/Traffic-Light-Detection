```python
import matplotlib.pyplot as plt
import numpy as np

# Confusion matrix values
confusion_matrix = np.array([[335, 530],
                             [0, 14]])

# Extract values from the confusion matrix
true_negative, false_positive = confusion_matrix[0]
false_negative, true_positive = confusion_matrix[1]

# Plot the bar chart
bar_width = 0.4
index = np.arange(2)

plt.bar(index, [true_negative, false_positive], bar_width, label='Class 0')
plt.bar(index, [false_negative, true_positive], bar_width, bottom=[true_negative, false_positive], label='Class 1')

plt.xlabel('Actual Classes')
plt.ylabel('Counts')
plt.title('Bar Chart for Anchor Box Applied Images')
plt.xticks(index, ['0', '1'])
plt.legend()

plt.show()
```
